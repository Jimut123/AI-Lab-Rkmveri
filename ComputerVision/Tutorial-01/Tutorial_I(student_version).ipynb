```python
#Displaying the Blue Channel
img_blue = img1[:,:,0].copy()
plt.imshow(img_blue,cmap = 'gray')                        #Displaying the Noisy-Image.
plt.show()

#Create and Display Noise
noise_blue = np.zeros((h,w), np.uint8)                 #Taking a Zero Matrix of the Image shape.
cv2.randn(noise_blue, 0, 100)                           #Defining the Noise with 0-Mean and 30-STD.
plt.imshow(noise_blue,cmap='gray')                     #Displaying the Noise.
plt.show()

#Adding Noise
img_blue_noisy = img_blue + noise_blue                    #Adding Noise to the Image.
plt.imshow(img_blue_noisy,cmap = 'gray')                  #Displaying the Noisy-Image.
plt.show()
```

**4. Add Noise to Specific Portions of an Image.**

EXAMPLE

```python
#Loading and Displaying the Image.
img = cv2.imread("/content/Supporting_Images/aamir_khan_photo.jpg")
img_copy = img.copy()
plt.imshow(img_copy[:,:,::-1])
plt.show()

#Drawing Bounding Box and Displaying the Image.
(xmin, ymin) = (129, 28)
(xmax, ymax) = (260, 159)
cv2.rectangle(img_copy,(xmin, ymin),(xmax, ymax),(255,255,0),8)
plt.imshow(img_copy[:,:,::-1])
plt.show()

#Target : Add Noise to Face.
#Step 1: Define Noise of the Same Dimension.
#Step 2: Add Noise to the Face

#Step 1:
(h,w) = (ymax-ymin, xmax-xmin)
noise = np.zeros((h,w), np.uint8)                  #Taking a Zero Matrix of the Image shape.
cv2.randn(noise, 0, 100)                           #Defining the Noise with 0-Mean and 30-STD.
plt.imshow(noise,cmap='gray')                      #Displaying the Noise.
plt.show()

#Step 2:
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')                      #Displaying the Gray Scale-Image.
plt.show()
img_gray[28:159, 129:260] = img_gray[28:159, 129:260] + noise
plt.imshow(img_gray, cmap='gray')                      #Displaying the Noisy-Image.
plt.show()
```

**UNIT-VII Draw Checker Board**

```python
img = np.ones((800,800,3), np.uint8)*255
black = np.zeros((100,100,3), np.uint8)
plt.imshow(img[:,:,::-1])
plt.show()
print("The shape of background is : ",img.shape)
plt.imshow(black)
plt.show()
print("The shape of black box is : ",black.shape)
(xmin, ymin, xmax, ymax) = (100, 0, 200,100)
c = 0
for i in range(8):
  for j in range(8):
    if(c%2==0):
      img[ymin:ymax,xmin:xmax,:] = black
      #plt.imshow(img[:,:,::-1])
      #plt.show()
    xmin+=100
    xmax+=100
    c+=1
  ymin+=100
  ymax+=100
  if(i%2!=0):
    xmin = 100
    xmax = 200
  else:
    xmin = 0
    xmax = 100
plt.imshow(img)
plt.title("Checker Board")
plt.show()
```
