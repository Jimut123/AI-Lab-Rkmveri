```python
from numpy import random as r
I = r.rand(256,256)
```

```python
I
```

```python
I[50:100,50:100] = 0
I[150:200,150:200] = 1
```

```python
from matplotlib import pyplot as plt
plt.imshow(I)
plt.show()
```

```python
plt.imshow(I,cmap='gray')
plt.show()
```

What will happen if we multiply the matrix by 0.5?

```python
I = r.rand(256,256) * 0.5;
plt.imshow(I,cmap='gray')
plt.show()
```

Next issue:
matplotlib will normalize input
e.g., map the lowest value to 0, and the highest to 1.Even though we multiplied by 0.5, 
the overall brightness is the same.


```python
I = r.rand(256,256) * 0.5
plt.imshow(I,cmap='gray',vmin=0.0,vmax=1.0)
plt.show()
```
