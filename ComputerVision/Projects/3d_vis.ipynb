```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = [1, 3, 5, 8, 6, 7, 1, 2, 4, 5]
y = [3, 4, 3, 6, 5, 3, 1, 2, 3, 8]

hist, xedges, yedges = np.histogram2d(x, y, bins=(4,4))
xpos, ypos = np.meshgrid(xedges[:-1]+xedges[1:], yedges[:-1]+yedges[1:])

xpos = xpos.flatten()/2.
ypos = ypos.flatten()/2.
zpos = np.zeros_like (xpos)

dx = xedges [1] - xedges [0]
dy = yedges [1] - yedges [0]
dz = hist.flatten()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='b', zsort='average')
plt.xlabel ("X")
plt.ylabel ("Y")

plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.array([0, 2, 5, 10, 2, 3, 5, 2, 8, 10, 11])
y = np.array([0, 2, 5, 10, 6, 4, 2, 2, 5, 10, 11])
# This example actually counts the number of unique elements.
binsOne = sorted(set(x))
binsTwo = sorted(set(y))
# Just change binsOne and binsTwo to lists.
hist, xedges, yedges = np.histogram2d(x, y, bins=[binsOne, binsTwo])

# The start of each bucket.
xpos, ypos = np.meshgrid(xedges[:-1], yedges[:-1])

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros_like(xpos)

# The width of each bucket.
dx, dy = np.meshgrid(xedges[1:] - xedges[:-1], yedges[1:] - yedges[:-1])

dx = dx.flatten()
dy = dy.flatten()
dz = hist.flatten()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='b', zsort='average')
```

```
import cv2
```

```
cutoff_frequency = 5
Gaussian_X = cv2.getGaussianKernel(ksize=cutoff_frequency*4+1,sigma=cutoff_frequency)
Gaussian_filter = np.dot(Gaussian_X, Gaussian_X.T)
print("The shape of the Gaussian Filter is : ",Gaussian_filter.shape)
plt.figure(figsize=(4,4))
plt.imshow(Gaussian_filter)
plt.show()
```

```
#A = np.array([[,1],[0,1]])
x=[]
for i in range(int(Gaussian_filter.shape[0])):
  x.append(i) 
y = []
for i in range(int(Gaussian_filter.shape[1])):
  y.append(i) 
```

```
x
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x, y, Gaussian_filter[x][y], 6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, Gaussian_filter[x][y], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

```
filter = Gaussian_filter.copy()
```

```
h=-filter
h[int(filter.shape[0]/2),int(filter.shape[0]/2)] = filter[int(filter.shape[0]/2),int(filter.shape[0]/2)]
h[int(filter.shape[0]/2),int(filter.shape[0]/2)] = h[int(filter.shape[0]/2),int(filter.shape[0]/2)] -h.sum()
```

```
x=[]
for i in range(int(h.shape[0])):
  x.append(i) 
y = []
for i in range(int(h.shape[1])):
  y.append(i) 
```

```

plt.figure(figsize=(4,4))
plt.imshow(h)
plt.show()
```

```
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, h[x][y], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```
