```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('test_img.jpg', cv2.IMREAD_COLOR)
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
kernel = np.ones((10,10),np.float32)/100
dst = cv2.filter2D(img1,-1,kernel)

plt.figure(figsize=(16, 16))
plt.subplot(121),plt.imshow(img1),plt.title('Original')
plt.subplot(122),plt.imshow(dst),plt.title('smooth')
plt.show()
```

```python
noiseReducedGaussianBlur = cv2.GaussianBlur(img1,(5,5),0)
plt.imshow(noiseReducedGaussianBlur[:,:,::-1], cmap='gray'),plt.title('Noise Reduced')
```

```python
cutoff_frequency = 100
filterX = cv2.getGaussianKernel(ksize=cutoff_frequency*4+1,
                               sigma=cutoff_frequency)
filterY = filterX.copy()

filter = np.dot(filterX, filterX.T)

# let's take a look at the filter!
plt.figure(figsize=(4,4))
plt.imshow(filter)
plt.show()
```

```python
blur = cv2.filter2D(img1, -1, filter)
plt.figure(figsize=(8, 8))
plt.imshow((blur).astype(np.uint8),cmap='gray')
plt.show()
```
